In additional to the raw polling data the polling client has a few helper functions to aid analysis, as shown in this notebook.

```python
# First load the client
from pollingclient.client import PollingClient
pc = PollingClient()
```

```python
pc.list_parties()
```

```python
from datetime import datetime
pc.party_in_power("Con", datetime(2017, 3, 6))
```

```python
pc.list_leaders()
```

```python
pc.get_leader_ratings("Clegg")
```

Leader rating progressions can be compared as below.

```python
pc.compare_ratings(["Clegg", "Kennedy", "Ashdown"], "Net", True)
```

```python
pc.list_pollsters()
```

The monthly_average method is a simple average of all poll results from that month, it doesn't attempt to adjust for over-representation of particular pollsters.

```python
pc.monthly_average().round(1)
```
