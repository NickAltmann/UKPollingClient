This notebook illustrates the basic use of the polling client.  First we import the library and create a client.

```python
from pollingclient.client import PollingClient
pc = PollingClient()
```

There are four pieces of data available through the client:
    - Party polling history.
    - Leader rating from Ipsos Mori.
    - An indication of when parties assumed power.
    - General election results for Great Britain (i.e. excluding Northern Ireland)
    
The attributes below return these datasets as pandas dataframes.  The first time you run then they will fetch the data from the web so will take short while.
